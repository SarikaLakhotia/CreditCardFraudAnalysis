```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

### Importing data

```python
data = pd.read_csv("C:\\Users\\ADMIN\\OneDrive\\Desktop\\Analysis Projects\\Credid Card Fraud Analysis ML\\creditcard.csv")
```

```python
print(data.head())
print(data.info())
```

```python
sns.countplot(x='Class', data=data)
plt.title('Class Distribution (Fraud vs Non-Fraud)')
plt.show()          
```

```python
features = ['Amount', 'Time']
X = data[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Train Isolation Forest Model

```python
model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
data['Anomaly'] = model.fit_predict(X_scaled)
```

```python
data['Anomaly'] = data['Anomaly'].map({1:0, -1:1})  #1 for anomlies and 0 for normal transactions
```

```python
data.to_csv('anamoly_results.csv', index = False)  #to save file for powerbi
```

### Visualise Anomalies

```python
plt.figure(figsize = (10,6))
anomalies = data[data['Anomaly'] == 1]
normal = data[data['Anomaly'] == 0]
plt.scatter(normal['Time'], normal['Amount'], c='blue', label='Normal', alpha=0.5)
plt.scatter(anomalies['Time'], anomalies['Amount'], c='red', label='Anomalies', alpha=0.5)
plt.title('Anomalies in Financial Transactions')
plt.xlabel('Time')
plt.ylabel('Transaction Amount')
plt.legend()
plt.show()
```

```python
if 'Class' in data.columns:
    print(confusion_matrix(data['Class'], data['Anomaly']))
    print(classification_report(data['Class'], data['Anomaly']))
```
